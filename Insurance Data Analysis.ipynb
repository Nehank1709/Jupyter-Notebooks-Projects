### 4) Explore the relationship between the feature and target column using a count plot of categorical columns and a scatter plot of numerical columns

#### Count plot for categorical columns

```python
plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
sns.countplot(x='sex', data=data)
plt.title('Count of Sex')

plt.subplot(1,3,2)
sns.countplot(x='smoker', data=data)
plt.title('Count of Smoker')

plt.subplot(1,3,3)
sns.countplot(x='region', data=data)
plt.title('Count of Region')

plt.xticks(rotation=45)
plt.show()
```

#### Observation: Count of Sex- Male count is slightly more than female count, Count of Smoker - Majority of participants are non smoker, Count of Region - Participants mostly belong to all 4 regions with slightly more to southeast region

#### Scatter plot of numerical columns w.r.t. insurance charges

```python
plt.figure(figsize=(15,15))
plt.subplot(3,1,1)
plt.title('age vs charges')
plt.xlabel('age')
plt.ylabel('charges')
plt.scatter(x='age', y='charges', data=data)

plt.subplot(3,1,2)
plt.title('bmi vs charges')
plt.xlabel('bmi')
plt.ylabel('charges')
plt.scatter(data['bmi'], data['charges'])

plt.subplot(3,1,3)
plt.title('children vs charges')
plt.xlabel('children')
plt.ylabel('charges')
plt.scatter(data['children'], data['charges'])


plt.show()
```

#### Observation: Age vs Charges - There is positive relationship between age and charges, as age increases, charges also increase on an average. Bmi vs Charges - There is positive relationship between BMI and charges, as BMI increases, charges also increase.  Children vs Charges - There is negative relationship between children and charges, as children count increases, charges decrease.
#### Hence, we can conclude that with increasing age and bmi( representor of health, high bmi means poor health), insurance charges increase and with more children count (can be due to high overall expenditure and less money left for insurance charge), insurance charges decrease.

```python
data1 = pd.read_csv('Insurance.csv')
data=data1.copy(deep=True)
```

```python
#feature engineering
data['sex'] = data['sex'].map({'female':0, 'male':1})
data['smoker'] = data['smoker'].map({'yes':8, 'no':9})
data['region'] = data['region'].map({'northeast':2, 'northwest':3, 'southeast':4, 'southwest': 5})
data.head()
```

```python
data.drop(['sex', 'region'], inplace=True, axis=1)
data.head()
```

```python
sns.heatmap(data1.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
x= data.loc[:,['age', 'bmi', 'children', 'smoker']]
y = data['charges']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
model=LinearRegression()
model.fit(x_train, y_train)
# print(x_test.head())
# print(y_test.head())

accuracy =model.score(x_test, y_test)
result= model.predict(x_test)
print(f'Model Accuracy: {accuracy*100 :.2f} %')
x_test['Charges Actual'] = y_test
x_test['Charges Predicted'] = result
x_test.reset_index().head(20)
```
